```python
from google.cloud import bigquery
import db_dtypes

# Initialize a client
client = bigquery.Client()

# Define your query
sql = """
-- Step 1: Identify hypotensive episodes
WITH HypotensiveEpisodes AS (
    SELECT
        ce.subject_id,
        ce.hadm_id,
        MIN(ce.charttime) as episode_start,
        MAX(ce.charttime) as episode_end
    FROM `physionet-data.mimiciv_icu.chartevents` as ce
    WHERE ce.itemid IN (220050, 220051, 220052)
    AND ce.valuenum < 90
    GROUP BY ce.subject_id, ce.hadm_id
    HAVING TIMESTAMP_DIFF(MAX(ce.charttime), MIN(ce.charttime), MINUTE) > 60
)

-- Step 4: Check for vasoactive drug administration
, VasoactiveDrugAdmin AS (
    SELECT
        dr.subject_id,
        dr.hadm_id
    FROM `physionet-data.mimiciv_hosp.prescriptions` as dr
    JOIN HypotensiveEpisodes he ON dr.subject_id = he.subject_id AND dr.hadm_id = he.hadm_id
    WHERE dr.drug IN ('Norepinephrine', 'Dopamine', 'Epinephrine', 'Vasopressin', 'Dobutamine', 'Phenylephrine')
    AND dr.starttime BETWEEN he.episode_start AND TIMESTAMP_ADD(he.episode_end, INTERVAL 1 HOUR)
)

-- Final Query
SELECT
    a.*,
    CASE
        WHEN d.icd_code LIKE 'I95%' OR d.icd_code LIKE 'R57%' THEN 1
        ELSE 0
    END AS Hypotension_ICD_at_discharge
FROM `physionet-data.mimiciv_hosp.admissions` as a
JOIN HypotensiveEpisodes he ON a.subject_id = he.subject_id AND a.hadm_id = he.hadm_id
JOIN VasoactiveDrugAdmin vda ON a.subject_id = vda.subject_id AND a.hadm_id = vda.hadm_id
LEFT JOIN `physionet-data.mimiciv_hosp.diagnoses_icd` as d ON a.hadm_id = d.hadm_id

"""

# Run the query and get the results
results = client.query(sql).result()

# Convert results to a pandas DataFrame for easier manipulation
import pandas as pd
df = results.to_dataframe()

print(df)
```
